```
import numpy as np
from audiolazy.lazy_lpc import *
import os

#mydata = np.loadtxt('001_SP_CPVV_010102G5.dat')
#print(len(mydata))


mydata_x = []
mydata_y = []
path = '/content/drive/My Drive/Machine Learning/BigDataProject/dataset/terremoti/'
fileList = os.listdir(path)
t = 0
for i in fileList:
    t= t+1
    loaded_file = np.loadtxt(path + i)
    print(t)
    mydata_x.append(lpc(loaded_file[0:300],order=6).numerator)
    for j in range(1,8):
      mydata_x[-1] = mydata_x[-1] + lpc(loaded_file[(j)*267:(j)*267 + 300],order=6).numerator
    mydata_y.append(0)

    somma = 0
    for k in range(0,24):
      maximum = np.amax(loaded_file[k*100:(k*100)+100])
      minimum = np.amin(loaded_file[k*100:(k*100)+100])
      somma = somma + (maximum - minimum)
      mydata_x[-1] = mydata_x[-1] + [((maximum - minimum)*24)/somma]
  
```

```
path = '/content/drive/My Drive/Machine Learning/BigDataProject/dataset/tuoni/'
fileList = os.listdir(path)
t = 0
for i in fileList:
    t= t+1
    loaded_file = np.loadtxt(path + i)
    print(t)
    mydata_x.append(lpc(loaded_file[0:300],order=6).numerator)
    for j in range(1,8):
      mydata_x[-1] = mydata_x[-1] + lpc(loaded_file[(j)*267:(j)*267 + 300],order=6).numerator
    mydata_y.append(2)

    somma = 0
    for k in range(0,24):
      maximum = np.amax(loaded_file[k*100:(k*100)+100])
      minimum = np.amin(loaded_file[k*100:(k*100)+100])
      somma = somma + (maximum - minimum)
      mydata_x[-1] = mydata_x[-1] + [((maximum - minimum)*24)/somma]
```

```
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import load_digits
from sklearn.metrics import accuracy_score

seed = 56

X_train, X_test, y_train, y_test = train_test_split(
    mydata_x, mydata_y, test_size=0.2, random_state=seed,stratify=mydata_y)

print("Training set size: ",len(X_train))
print("Test set size: ",len(X_test))
print("Training set size: ",len(y_train))
print("Test set size: ",len(y_test))
```

```
# Definition
mlp = MLPClassifier(
    hidden_layer_sizes=(42),
    batch_size='auto', 
    learning_rate_init=0.001,
    momentum=0.9,  
    max_iter=200,
    early_stopping=True, 
    validation_fraction=0.1, 
    n_iter_no_change=30,
    verbose=True
  )

# Start training
mlp.fit(X_train,y_train)
```

```
# Early stopping parameters
early_stopping = False
validation_fraction = 0.2
evaluation_epochs = 2 # validation evaluation rate (in terms of number of epochs)
max_iter = evaluation_epochs*80*4 # max_iter needs to be a multiple of evaluation_epochs
n_iter_no_change=5

# Validation split
X_train_1, X_val, y_train_1, y_val = train_test_split(
    X_train, y_train, test_size=validation_fraction, random_state=seed) #divido il training set di prima in training e validation set (10%)

# Model definition without early stopping
mlp = MLPClassifier(
    hidden_layer_sizes=(42),# 1000 neuroni primo livello nascosto, 10 neuroni nel secondo livello nascosto
    batch_size='auto',
    learning_rate_init=0.001,
    momentum=0.9,  
    max_iter=evaluation_epochs,
    early_stopping=False,
    verbose=False
)

# Curve lists
t_curve = []
v_curve = []
```

```
# Training
for i in range(int(max_iter/evaluation_epochs)):
  mlp.partial_fit(X_train_1,y_train_1, classes=np.unique(mydata_y))

  # Prediction
  y_pred = mlp.predict(X_train_1)
  yv_pred = mlp.predict(X_val)

  # Append the current performance to the curve
  t_curve.append(accuracy_score(y_train_1,y_pred))
  v_curve.append(accuracy_score(y_val,yv_pred))
  print("Epoch ",i*evaluation_epochs+1,'\tTraining Accuracy ',t_curve[-1],'\tValidation Accuracy ',v_curve[-1])

  # Early stopping
  if early_stopping and len(v_curve) > n_iter_no_change:
    if max(v_curve[-n_iter_no_change:]) < v_curve[-n_iter_no_change-1]:
      print("Early stopping - Epoch ",i*evaluation_epochs+1)
      break
```

```
t = [i+1 for i in range(len(v_curve))]
plt.plot(t,t_curve,t,v_curve)
plt.ylabel('Accuracy')
plt.xlabel('N of Epochs')
plt.title('Learning Curve')
plt.legend(["Training Accuracy","Validation Accuracy"])
plt.show()
```
